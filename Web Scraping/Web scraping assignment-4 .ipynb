**Q1) Scrape the details of most viewed videos on YouTube from Wikipedia.<br>

You need to find following details:<br>
A) Rank<br>

C) Artist<br>
D) Upload date<br>
E) Views**<br>

```python
#Connecting to the webdriver and opening the wikipedia
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#scraping the rank of the video  
rank=[]
try:
    ranks=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in ranks:
        rank.append(i.text)
except NoSuchElementException:#handling no such element exception
    rank.append("-")
    
#scraping the name of the video 
name=[]
try:
    names=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in names:
        name.append(i.text)
except NoSuchElementException:#handling no such element exception
    name.append('-')
        
#scraping the artist of the video 
artist=[]
try:
    artists=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artists:
        artist.append(i.text)
except NoSuchElementException:#handling no such element exception
    artist.append('-')
    
#scraping the upload date of the video     
upload_date=[]
try:
    upload_date_list=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in upload_date_list:
        upload_date.append(i.text)
except NoSuchElementException:#handling no such element exception
    upload_date.append('-')
    
#scraping the views of the video     
views=[]
try:
    views_list=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in views_list:
        views.append(i.text)
except NoSuchElementException:#handling no such element exception
    views.append('-')  

#creating the DataFrame
df=pd.DataFrame({})
df["Rank"]=rank
df['Name']=name
df['Artist']=artist
df["Upload_date"]=upload_date
df["Views(in billions)"]=views
df    
```

```python
#scraping the repository title
repo_title=[]
try:
    title=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
    for i in title:
        repo_title.append(i.text)
except NoSuchElementException:#handling no such element exception
    repo_title.append("-") 

#scraping the language used     
language_used=[]
try:
    lang=driver.find_elements_by_xpath('//div[@class="f6 color-fg-muted mt-2"]/span[1]')
    for i in lang:
        language_used.append(i.text)
except NoSuchElementException:#handling no such element exception
    language_used.append("-") 
    
#scraping the repository description  
repo_desc=[]
try:
    desc=driver.find_elements_by_xpath('//p[@class="col-9 color-fg-muted my-1 pr-4"]')
    for i in desc:
        repo_desc.append(i.text)
except NoSuchElementException:#handling no such element exception
    repo_desc.append("-")
    
#scraping the contributors count     
Contributors_count=[]
Contributors_url=[]

i=1

while True:
    try:
        Contributors_url.append(driver.find_element_by_xpath(f'//div//article[@class="Box-row"][{i}]//h1[@class="h3 lh-condensed"]//a').get_attribute('href'))
        i+=1
    except:
       
       break
        
for i in Contributors_url:
    driver.get(i)
    time.sleep(2)
    try:
        Contributors_count.append(driver.find_element_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]').text.split('Contributors')[1].split('\n')[0])     
    except:
        Contributors_count.append('-')    
```
